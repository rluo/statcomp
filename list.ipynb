```r
library(microbenchmark)
```

```r
n <- 1000
set.seed(100)

matrix_data <- matrix(rnorm(n^2), nrow = n, ncol = n)
data_frame_data <- data.frame(matrix_data)
```

```r
bench <- microbenchmark(matrix_data * 10, data_frame_data*10)
bench
plot(bench)
```

```r
bench <- microbenchmark(matrix_data %*% matrix_data,  as.matrix(data_frame_data) %*% as.matrix(data_frame_data) )
bench
plot(bench)
```
